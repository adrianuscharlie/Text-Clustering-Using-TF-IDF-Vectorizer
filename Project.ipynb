```python
#Import pandas untuk membuat dataframe dari list judul berita
import pandas as pd
```

```python
#Reading file kumpulan judul yang sudah disimpan dalam format csv
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Project/Bahan_lengkap/Bahan - Sheet1.csv')
df
```

```python
#Mengubah dataframe menjadi list agar bisa mengolah data
text=df.values.tolist()
text
```

```python
#Membuat function untuk preprocessing mengambil judul berita
def Title(text):
  for i in range(len(text)):
    temp=str(text[i])
    text[i]=temp[52:temp.index("http")-1]
  return text
```

```python
#Menjalankan function Title untuk mengambil judul berita
text=Title(text)
text
```

```python
#Menginstal package PySastrawi
!pip install PySastrawi
```

```python
#melakukan import modul StemmerFactory dan StopWordRemoverFactory untuk menghilangkan stopword
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
#Menghilangkan stopword
stemmer = StemmerFactory().create_stemmer()
stopwords = StopWordRemoverFactory().get_stop_words()
```

```python
#Membuat Data Frame dari list judul 
columns=['Sentences']
df_new=pd.DataFrame(text,columns=columns)
df_new
```

```python
#Melakukan Pre-Processing dengan metode stemmming dan stopword removal
def text_preprocess(series, stemmer, stopwords):
    df = series.str.replace("\n\t",  " ")
    df = df.str.replace(r"[^a-zA-Z ]+", "")
    df = df.str.lower()
    df = df.apply(lambda x: ' '.join([stemmer.stem(item) for item in x.split() if item not in stopwords]))
    return df
```

```python
#Menampilkan DataFrame yang berisi sentences/judul sebelum dan sesudah dilakukan preprocessing
df_new['processed_text'] = text_preprocess(df_new['Sentences'], stemmer, stopwords)
df_new
```

```python
#Mengambil text hasil preprocessing dari data frame
corpus=df_new['processed_text'].values.tolist()
corpus
```

```python
#Membuat wordbank yang akan digunakan dalam melakukan clustering
feature=['cancer', 'mental','heart','diabetes','hiv', 'stroke']
```

```python
#Melakukan import TfidfVectorizer yang berguna untuk melakukan perhitungan tfidf dari dokumen tersebut
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(stop_words='english',max_features=6,vocabulary=feature)
```

```python
#Transform text menjadi sebuah vector
vector=vectorizer.fit_transform(corpus)
print(vector)
```

```python
#Mencetak feature name dari tfidf 
feature_names=vectorizer.get_feature_names()
feature_names
```

```python
#Mengubah vector diatas menjadi list agar bisa dimasukan ke dalam data frame
dense=vector.todense()
dense_list=dense.tolist()
dense_list
```

```python
#Menampilkan data frame berisi nilai tf-idf
tfidf=pd.DataFrame(dense_list,columns=feature_names)
tfidf
```

```python
#Melakukan import modul KMeans untuk dapat melakukan clustering
from sklearn.cluster import KMeans
```

```python
#Membuat model KMeans clustering dari tfidf
k=6
model=KMeans(n_clusters=k,init='k-means++',max_iter=1000,n_init=1)
model.fit(tfidf)
```

```python
#Menampilkan data frame berisi text dan label hasil clusternya
df_new['Cluster']=model.labels_
df_new
```

```python
#Melakukan export data frame menjadi excel agar dapat dianalisa
df_new.to_excel("cluster_new.xls")
```
